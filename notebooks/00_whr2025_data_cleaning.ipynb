This jupyter notebook is the cleaning process of the whr 2025 dataset

```python
# Import libraries
import pandas as pd
```

```python
# Read in world happiness report data file
whr_df = pd.read_excel('../data_raw/WHR25_Data_Figure_2.1v3.xlsx')

whr_df.head()
```

```python
# Check dataframe info
print(whr_df.columns)
print(whr_df.info())
```

```python
# Standardize column names
whr_df.columns = (
    whr_df.columns
    .str.strip()                 # remove stray spaces
    .str.lower()                 # lowercase for consistency
    .str.replace(' ', '_')       # replace spaces with underscores
    .str.replace(':', '')        # remove colons
    .str.replace('(', '')        # remove parentheses
    .str.replace(')', '')        # remove parentheses
    .str.replace('dystopia_+_residual', 'dystopia_plus_residual')  # fix column name
)

# Check result
whr_df
```

```python
# Check for dupes
duplicates = whr_df[whr_df.duplicated(subset=['country_name', 'year'], keep=False)]
duplicates
```

This part of the Jupyter Notebook is the reading in of wdi country data to compare and clean country names.

```python
# Read in wb wdi country data to compare country_name columns
wb_wdi_country_df = pd.read_csv('..\data\wb_wdi_country_df_final.csv')
wb_wdi_country_df
```

```python
# Check differences in country name column between dataframes
set(whr_df['country_name']) - set(wb_wdi_country_df['country_name'])
```

```python
# Build a rename dictionary for whr to match wdi
rename_dict = {
    'Congo': 'Congo, Rep.',
    'DR Congo': 'Congo, Dem. Rep.',
    'CÃ´te dâ€™Ivoire': "Cote d'Ivoire",
    'Egypt': 'Egypt, Arab Rep.',
    'Gambia': 'Gambia, The',
    'Hong Kong SAR of China': 'Hong Kong SAR, China',
    'Iran': 'Iran, Islamic Rep.',
    'Kyrgyzstan': 'Kyrgyz Republic',
    'Republic of Korea': 'Korea, Rep.',
    'Republic of Moldova': 'Moldova',
    'Slovakia': 'Slovak Republic',
    'Somalia': 'Somalia, Fed. Rep.',
    'Swaziland': 'Eswatini',
    'Syria': 'Syrian Arab Republic',
    'Turkey': 'Turkiye',
    'TÃ¼rkiye': 'Turkiye',
    'Venezuela': 'Venezuela, RB',
    'Yemen': 'Yemen, Rep.',
    'State of Palestine': 'West Bank and Gaza'
}

# Apply dictionary
whr_df['country_name'] = whr_df['country_name'].replace(rename_dict)
```

```python
# Rerun this to check differences left after rename dictionary
set(whr_df['country_name']) - set(wb_wdi_country_df['country_name'])
```

```python
# Inspect data for Somaliland Region and Somalia, Fed. Rep. differences
whr_df[whr_df['country_name'].str.contains('Somali', case=False)]
```

```python
# Inspect data to look for any other year/rank discrepancies
(
    whr_df.duplicated(subset=['year', 'rank'], keep=False)
    .groupby(whr_df['year'])
    .sum()
)
```

```python
# Find all rows where year/rank combo appears more than once
dup_ranks = whr_df[whr_df.duplicated(subset=['year', 'rank'], keep=False)]

# Check rows with duplicate year/rank
dup_ranks = dup_ranks.sort_values(['year', 'rank']).reset_index(drop=True)
dup_ranks[['year', 'rank', 'country_name']].head(10)
```

```python
# Pick final preferred names (mix of WHR/WDI)
# Chose to combine Somalia, Fed. Rep. and Somaliland Region for data consistency (both as Somalia).
final_names = {
    'Congo, Rep.': 'Congo',
    'Congo, Dem. Rep.': 'DR Congo',
    'Egypt, Arab Rep.': 'Egypt',
    'Gambia, The': 'Gambia',
    'Hong Kong SAR, China': 'Hong Kong SAR of China',
    'Iran, Islamic Rep.': 'Iran',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Korea, Rep.': 'Republic of Korea',
    'Slovak Republic': 'Slovakia',
    'Somalia, Fed. Rep.': 'Somalia',
    'Somaliland Region': 'Somalia',
    'Syrian Arab Republic': 'Syria',
    'Turkey': 'Turkiye',
    'TÃ¼rkiye': 'Turkiye',
    'Venezuela, RB': 'Venezuela',
    'Yemen, Rep.': 'Yemen'
}

whr_df['country_name_final'] = whr_df['country_name'].replace(final_names)

# Check result
whr_df.head()
```

```python
# See original vs final country name
whr_df[['country_name', 'country_name_final']]
whr_df
```

```python
# Strip country_name_final to remove whitespace before merge with wdi data (precautionary)
whr_df['country_name_final'] = whr_df['country_name_final'].str.strip()

# Keep a copy of the original rank for reference
whr_df['original_rank'] = whr_df['rank']

# Check result
whr_df
```

```python
# Save whr data with renamed countries as csv (before Somalia 2015 merge and country reranking)
whr_df.to_csv(r'..\data_clean\whr_df_country_rename.csv', index=False)
print('whr_df_country_rename.csv has been saved')
```

```python
# Drop rank before aggregation
whr_df_rnk_col = whr_df.drop(columns=['rank'], errors='ignore')

# Check result
whr_df_rnk_col
```

```python
# Group by year and country_name_final - averaging numeric columns
whr_df_final = (
    whr_df_rnk_col
    .groupby(['year', 'country_name_final'], as_index=False)
    .mean(numeric_only=True)
)

# Check result
whr_df_final
```

```python
# Recalculate rank based on life evaluation
whr_df_final['rank'] = whr_df_final.groupby('year')['life_evaluation_3-year_average'] \
    .rank(ascending=False, method='min')

# Convert rank columns back to int
whr_df_final['rank'] = whr_df_final['rank'].astype('int')
whr_df_final['original_rank'] = whr_df_final['original_rank'].astype('int')

# Check result
whr_df_final
```

```python
# Sort by country, then by year
whr_df_final = whr_df_final.sort_values(by=['country_name_final', 'year']).reset_index(drop=True)

# Preview
print(whr_df_final[['country_name_final', 'year', 'life_evaluation_3-year_average', 'rank']].head(15))
```

```python
# Check for duplicates in year + country_name_final
dupes = whr_df_final[whr_df_final.duplicated(subset=['year', 'country_name_final'], keep=False)]

if duplicates.empty:
    print("âœ… All country-year combinations are unique.")
else:
    print("âš ï¸ Duplicates found:")
    print(dupes)

# Check for 2015 Somalia merged correctly
whr_df_final[whr_df_final['country_name_final'] == 'Somalia']
```

```python
# Rename country_name_final to country_name
whr_df_final = whr_df_final.rename(columns={'country_name_final': 'country_name'})

# Verify
whr_df_final.columns
```

```python
# Reorder and drop columns for final whr dataframe
final_columns = ['year', 'country_name', 'rank', 'life_evaluation_3-year_average', 'explained_by_log_gdp_per_capita', 'explained_by_healthy_life_expectancy', 'explained_by_perceptions_of_corruption']

# Keep only these columns and reorder
whr_df_final = whr_df_final[final_columns]

# Preview the cleaned, ordered dataframe
whr_df_final
```

```python
# Round all float columns to 3 decimals
whr_df_final = whr_df_final.round(3)
whr_df_final
```

```python
# Save final whr data as csv before merge with wdi data
whr_df_final.to_csv(r'..\data\whr_df_final.csv', index=False)
print('whr_df_final.csv has been saved')
```
