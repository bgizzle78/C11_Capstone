This Jupyter Notebook is the cleaning process of the world bank databank world development (8) indicators csv.

```python
# Import libraries
import pandas as pd
```

```python
# Read in world bank wdi data file
wb_wdi_df_wide = pd.read_csv('../data_raw/WB_WDI_Data.csv')

# Check data
print(wb_wdi_df_wide.info())
wb_wdi_df_wide.head()
```

```python
# Check for missing values
wb_wdi_df_wide[wb_wdi_df_wide['Country Name'].isna() | wb_wdi_df_wide['Country Code'].isna()]
```

```python
# Drop last 5 rows (blank rows and metadata info)
wb_wdi_df_wide.drop(wb_wdi_df_wide.tail(5).index, inplace=True)
wb_wdi_df_wide.tail(6)
```

```python
# Check columns
wb_wdi_df_wide.columns
```

```python
# Melt the dataframe to long format
wb_wdi_df = wb_wdi_df_wide.melt(
    id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'],
    value_vars=[col for col in wb_wdi_df_wide.columns if 'YR' in col],
    var_name='Year',
    value_name='Value'
)

# Clean up the 'Year' column (remove '[YR2011]' part and change to int)
wb_wdi_df['Year'] = wb_wdi_df['Year'].str.extract('(\d{4})').astype(int)

# Convert 'Value' to numeric and round to 2 decimals
wb_wdi_df['Value'] = pd.to_numeric(wb_wdi_df['Value'], errors='coerce').round(2)

# Strip country name to remove whitespace
wb_wdi_df['Country Name'] = wb_wdi_df['Country Name'].str.strip()

# Check the result
wb_wdi_df.head()
```

```python
# # Standardize column names for consistency between data for cleaner merges
wb_wdi_df.rename(columns={'Country Name': 'country_name', 'Country Code': 'id', 'Series Name': 'series_name', 'Series Code': 'series_code', 'Year': 'year', 'Value': 'value'}, inplace=True)

# Check result
wb_wdi_df
```

```python
# Reorder columns for readability and drop unnecessary columns
column_order = ['id', 'country_name', 'year', 'value', 'series_name']
wb_wdi_df = wb_wdi_df[column_order]

# Check result
wb_wdi_df.head()
```

```python
# Save cleaned data as csv file
wb_wdi_df.to_csv(r'..\data_clean\wb_wdi_df_clean.csv', index=False)
print('wb_wdi_df_clean.csv has been saved')
```

This part of the Jupyter Notebook is the merging of country_data_clean.csv and wb_wdi_df_clean.csv

```python
# Read in country metadata csv
country_metadata = pd.read_csv('..\data_clean\country_metadata_clean.csv')

# Check result
country_metadata.head()
```

```python
# Check which country codes and country names match
print(set(wb_wdi_df['id']).difference(set(country_metadata['id'])))
print(set(wb_wdi_df['country_name']).difference(set(country_metadata['country_name'])))
```

```python
# Merging dataframes
wb_wdi_country_merged = pd.merge(wb_wdi_df, country_metadata, how='left', on=['id', 'country_name'])

# Check result
print(wb_wdi_country_merged.shape)
wb_wdi_country_merged.head()
```

```python
# Final clean of columns and column reorder
wb_wdi_country_final = wb_wdi_country_merged.drop(columns=['capital_city'])

wb_wdi_country_final = wb_wdi_country_final[['id', 'country_name', 'year', 'series_name', 'value', 'income_level', 'region', 'longitude', 'latitude', 'iso2Code']]

# Check result
print(wb_wdi_country_final.shape)
wb_wdi_country_final.head()
```

```python
# Save final merged data as csv
wb_wdi_country_final.to_csv(r'..\data\wb_wdi_country_df_final.csv', index=False)
print('wb_wdi_country_df_final.csv has been saved')
```
