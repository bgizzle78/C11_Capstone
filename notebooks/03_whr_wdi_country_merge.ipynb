This Jupyter Notebook is the merging of the whr data with the wb wdi data. Final merge to make usable data for visualizations.

```python
# Import libraries
import pandas as pd
```

```python
# Read in cleaned csv's for merge
wb_wdi_country_df = pd.read_csv('../data_clean/wb_wdi_country_df_final.csv')
whr_df = pd.read_csv('../data_clean/whr_df_final.csv')
print('CSVs read in successfully')
```

```python
# Dupes for wdi data
wdi_duplicates = wb_wdi_country_df[wb_wdi_country_df.duplicated(subset=['year', 'country_name', 'series_name'], keep=False)]
print('Number of duplicate rows in wdi data:', len(wdi_duplicates))
print(wdi_duplicates[['year', 'country_name', 'series_name']])

# Dupes for whr data
whr_duplicates = whr_df[whr_df.duplicated(subset=['year', 'country_name'], keep=False)]
print('Number of duplicate rows in whr data:', len(whr_duplicates))
print(whr_duplicates[['year', 'country_name']])
```

```python
# Check whr data
print(whr_df.info())
whr_df.head()
```

```python
# Rename columns in whr dataframe
whr_df_final = whr_df.rename(columns={
    'life_evaluation_3-year_average': 'Happiness Score',
    'explained_by_log_gdp_per_capita': 'GDP Contribution',
    'explained_by_healthy_life_expectancy': 'Life Expectancy Contribution',
    'explained_by_perceptions_of_corruption': 'Corruption Impact'
}, inplace=False)

# Check result
whr_df_final
```

```python
# Check wdi data
print(wb_wdi_country_df.info())
wb_wdi_country_df.head()
```

```python
# Rename columns in wdi dataframe
wdi_country_df_final = wb_wdi_country_df.rename(columns={
    'income_level': 'Income Level',
    'region': 'Region'
}, inplace=False)

# Check result
wdi_country_df_final
```

```python
# Perform a left join
left_merge = pd.merge(
    whr_df_final,
    wdi_country_df_final,
    on=['year', 'country_name'],
    how='left'
)

print("Left join shape:", left_merge.shape)
```
